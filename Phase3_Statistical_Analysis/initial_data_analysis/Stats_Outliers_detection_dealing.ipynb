```python
### median is always the second quartile
```

```python
q1=df['hp'].quantile(0.25)  ### 100 parts 
q3=df['hp'].quantile(0.75)
```

```python
q1
```

```python
q3
```

```python
### Inter Quartile Range (IQR )
```

```python
iqr=q3-q1
iqr
```

```python
### 1.5 IQR
```

```python
upper_bound=q3+1.5*iqr
lower_bound=q1-1.5*iqr
```

```python
upper_bound
```

```python
lower_bound
```

```python
filter=df['hp']>upper_bound
df[filter]
```

```python
filt=df['hp']<upper_bound
df2=df[filt]
df2
```
