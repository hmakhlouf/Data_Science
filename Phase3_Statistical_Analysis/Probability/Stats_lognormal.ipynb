```python
from scipy.stats import expon ### exponential 
```

```python
data=expon.rvs(scale=1,loc=0,size=1000)
data
```

```python
ax=sns.distplot(data,hist=True,kde=False)
ax.set(xlabel='Expo dist',ylabel='frequency')
```

```python
ax=sns.distplot(data,hist=True,kde=True)
ax.set(xlabel='Expo dist',ylabel='frequency')
```
